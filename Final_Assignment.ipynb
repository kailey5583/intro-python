```


#This is for the graph on economy  
import pandas as pd
import matplotlib.pyplot as pyplot
import numpy as np
file_name = "https://raw.githubusercontent.com/mlepinski/Python-Worksheets/master/Matplotlib/Data/covid_concern_polls_adjusted.csv"
covid_data= pandas.read_csv(file_name)
covid_data.set_index("enddate",inplace= True)
covid_data.head()
economy_concern= covid_data[covid_data["subject"]== "concern-economy"]
economy_concern.head()
economy_concern.rolling(25).mean().plot(y="very")
plt.xticks(rotation=45)
plt.xlabel("End Dates")
plt.ylabel("Very Concerned Respondents (Average)")
plt.title("Number of Very Concerned Americans over Time")
plt.show()
```
